```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['lines.linewidth'] = 2.5
plt.rcParams['lines.markersize'] = 7
plt.rcParams['lines.markeredgewidth'] = 2.5
```

```python
# Device
num_device_list = [3, 5, 7, 9, 11, 13]
gain_array_GNN = np.load('./save_model/save_results/GNN/discriminant_gains_device.npy', allow_pickle=True)
gain_array_cvx_with_ris = np.load('./save_model/save_results/discriminant_gain_device_with_ris.npy', allow_pickle=True)
gain_array_cvx_without_ris = np.load('./save_model/save_results/discriminant_gain_device_without_ris.npy', allow_pickle=True)
# gain_array_cvx_random = np.load('./save_model/save_results/discriminant_gain_device_random.npy', allow_pickle=True)
plt.plot(num_device_list, gain_array_GNN, marker='*', label='GNN_based', color='#8B0000')
plt.plot(num_device_list, gain_array_cvx_with_ris, marker='D', markerfacecolor='none', label='CVX_based_with_RIS', color='black')
plt.plot(num_device_list, gain_array_cvx_without_ris, marker='s', markerfacecolor='none', label='CVX_based_without_RIS', color='purple')
# plt.plot(num_device_list, gain_array_cvx_random, marker='o', label='Random', color='grey')
plt.xlabel('Number of Devices')
plt.ylabel('Discriminant Gain')
plt.legend()
plt.savefig('./save_model/save_results/graphs/device.pdf', format='pdf')
plt.show()
```

```python
# Power
power_list = [4, 5, 6, 7, 8, 9, 10, 11, 12]
gain_array_GNN = np.load('./save_model/save_results/GNN/discriminant_gains_power.npy', allow_pickle=True)
gain_array_cvx_with_ris = np.load('./save_model/save_results/discriminant_gain_power_with_ris.npy', allow_pickle=True)
gain_array_cvx_without_ris = np.load('./save_model/save_results/discriminant_gain_power_without_ris.npy', allow_pickle=True)
# gain_array_cvx_random = np.load('./save_model/save_results/discriminant_gain_power_random.npy', allow_pickle=True)
plt.plot(power_list, gain_array_GNN, marker='*', label='GNN_based', color='#8B0000')
plt.plot(power_list, gain_array_cvx_with_ris, marker='D', markerfacecolor='none', label='CVX_based_with_RIS', color='black')
plt.plot(power_list, gain_array_cvx_without_ris, marker='s', markerfacecolor='none', label='CVX_based_without_RIS', color='purple')
# plt.plot(power_list, gain_array_cvx_random, marker='o', label='Random', color='grey')
plt.xlabel('Value of Power bar')
plt.ylabel('Discriminant Gain')
plt.legend()
plt.savefig('./save_model/save_results/graphs/power.pdf', format='pdf')
plt.show()
```
