```python
num_device_list = [3, 5, 7, 9, 11, 13]
gain_array_GNN = np.load('./save_model/save_results/GNN/discriminant_gains_device.npy', allow_pickle=True)
gain_array_cvx = np.load('./save_model/save_results/discriminant_gain_with_num_device_ris.npy', allow_pickle=True)
gain_array_cvx_baseline = np.load('./save_model/save_results/discriminant_gain_with_num_device_baseline_ris.npy', allow_pickle=True)
plt.plot(num_device_list, gain_array_GNN, marker='o', label='GNN', color='red')
plt.plot(num_device_list, gain_array_cvx, marker='o', label='CVX', color='blue')
plt.plot(num_device_list, gain_array_cvx_baseline, marker='o', label='random', color='green')
plt.xlabel('Number of Devices')
plt.ylabel('Discriminant Gain')
plt.grid(True)
plt.legend()
plt.show()
```
